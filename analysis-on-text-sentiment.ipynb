```r
# drawing wordcloud with r: code borrowed from
# http://www.sthda.com/english/wiki/word-cloud-generator-in-r-one-killer-function-to-do-everything-you-need
# http://www.sthda.com/english/wiki/text-mining-and-word-cloud-fundamentals-in-r-5-simple-steps-you-should-know
# # Install
# install.packages("tm")  # for text mining
# install.packages("SnowballC") # for text stemming
# install.packages("wordcloud") # word-cloud generator
# install.packages("RColorBrewer") # color palettes
# Load
library(tm)
library(SnowballC)
library(wordcloud)
library(RColorBrewer)
library(tidyverse)
# source('http://www.sthda.com/upload/rquery_wordcloud.r')
select <- dplyr::select

# data import and clean -------------------------------------------------------------

# setwd("~/Dropbox/PhD/Modules/SPH6004/project 2")


breed_labels <- read_csv("../input/breed_labels.csv")
color_labels <- read_csv("../input/color_labels.csv")
state_labels <- read_csv("../input/state_labels.csv")

AdoptionSpeed_labels <- c("Same day as the listing",
                          "Btw 1 and 7 days",
                          "Btw 8 and 30 days",
                          "Btw 31 and 90 days",
                          "No adoption after 100 days")
table.label <- function(tb,breed_labels, color_labels, state_labels){
  tb %>% 
    mutate_at(vars(starts_with("Breed")),
              factor, 
              levels = breed_labels$BreedID,
              labels = breed_labels$BreedName) %>%
    mutate_at(vars(starts_with("Color")),
              factor, 
              levels = color_labels$ColorID,
              labels = color_labels$ColorName) %>%
    mutate(Gender = factor(Gender, levels = 1:3,
                           labels = c("Male", "Female", "Mixed")),
           Type = factor(Type, levels = 1:2,
                         labels = c("Dog", "Cat")),
           MaturitySize = factor(MaturitySize, levels = c(1:4, 0),
                                 labels = c("Small", "Medium", "Large",
                                            "Extra Large", "Not specified")),
           FurLength = factor(FurLength, levels = c(1:3, 0),
                              labels = c("Short", "Medium", "Long",
                                         "Not specified")),
           Vaccinated = factor(Vaccinated, levels = 1:3,
                               labels = c("Yes", "No", "Not sure")),
           Dewormed = factor(Dewormed, levels = 1:3,
                             labels = c("Yes", "No", "Not sure")),
           Sterilized = factor(Sterilized, levels = 1:3,
                               labels = c("Yes", "No", "Not sure")),
           Health = factor(Health, levels = c(1:3, 0),
                           labels = c("Healthy", "Minor Injury",
                                      "Serious Injury", "Not specified")),
           State = factor(State,
                          levels = state_labels$StateID,
                          labels = state_labels$StateName))
}
train <- read_csv("../input/train/train.csv") %>%
  table.label(.,breed_labels,color_labels, state_labels) %>% 
  mutate(AdoptionSpeed = factor(AdoptionSpeed,
                                levels = 0:4,
                                labels = AdoptionSpeed_labels))



test <- read_csv("../input/test/test.csv") %>%
  table.label(.,breed_labels,color_labels, state_labels) 


# wordcloud ---------------------------------------------

dat <- train %>% 
  select(Description)
text <- paste(dat, collapse=' ')
# png('wordcloud.png')
# res <- rquery.wordcloud(text, type='text', 
#                  lang="english", excludeWords = NULL, 
#                  textStemming = FALSE,  colorPalette="Dark2",
#                  max.words=100)
# dev.off()

# compare popular words across adoption speed --------------------------------------

library(tidytext)
data(stop_words)

df_text <- train %>% 
  select(text= Description, AdoptionSpeed) %>% 
  mutate(line=1:nrow(train)) %>% 
  unnest_tokens(word, text) %>% 
  anti_join(stop_words) %>% 
  anti_join(data.frame(word=as.character(1:100)))

tb_count <- df_text %>%
  group_by(AdoptionSpeed) %>% 
  count(word, sort = TRUE) %>% 
  spread(AdoptionSpeed, n) 
tb_count$Total <-  sapply(1:nrow(tb_count), function(i)sum(tb_count[i,-1], na.rm=T))
tb_count <- tb_count %>% 
  arrange(desc(Total)) 

# data for plotting
df_count <- tb_count %>% 
  gather(AdoptionSpeed, n, 2:6) %>% 
  mutate(prop= n/Total) %>% 
  filter(Total>1000) %>% 
  mutate(word = reorder(word, n), AdoptionSpeed=factor(AdoptionSpeed, levels=AdoptionSpeed_labels)) 

# plot word frequency by AdoptionSpeed
df_count %>% 
  ggplot(aes(word, n, group=AdoptionSpeed, fill=AdoptionSpeed)) +
  geom_col() +
  xlab(NULL) +
  coord_flip()

# plot word proportion by AdoptionSpeed
df_count %>% 
  ggplot(aes(word, prop, group=AdoptionSpeed, fill=AdoptionSpeed)) +
  geom_col() +
  xlab(NULL) +
  coord_flip()

# sentiment ---------------------------------------------------------------

lexi <- get_sentiments("nrc") 
table(lexi$sentiment)

df_count %>% 
  left_join(lexi, by='word') # unmatched words mostly don't have sentiments

tb_sent <- df_count %>% 
  left_join(lexi, by='word') %>% 
  select(AdoptionSpeed, n, sentiment) %>%
  filter(!is.na(sentiment)) %>% 
  group_by(AdoptionSpeed, sentiment) %>% 
  summarise(s=sum(n)) %>% 
  group_by(sentiment) %>% 
  mutate(p=s/sum(s))

tb_sent %>% 
  ggplot(aes(sentiment, p, group=AdoptionSpeed, fill=AdoptionSpeed)) +
  geom_col() +
  xlab(NULL) +
  coord_flip()

# sentiment dataset
tb_text <- df_text %>% 
  left_join(lexi, by='word') %>% 
  filter(!is.na(sentiment)) %>% 
  count(line, AdoptionSpeed, sentiment) %>% 
  spread(sentiment, n) # a lot of NAs
# replace NA with 0
tb_text[is.na(tb_text)] <- 0

# sparse matrix
train1 <- tb_text %>% ungroup 
train1_label <- as.numeric(train1[,2][[1]])-1
train1 <- as.matrix(train1[, -c(1:2)])

# # check availability of variables
# tb_text %>% 
#   ungroup %>% 
#   select(-c(1:2)) %>% 
#   summarise_all(function(x)sum(x!=0))

# # validate accuracy
# idx_train <- sample(1:nrow(train1), .8*nrow(train1))
# train2 <- train1[idx_train, ]
# train2_label <- train1_label[idx_train]
# test2 <- train1[-idx_train, ]
# test2_label <-  train1_label[-idx_train]

# modelling ---------------------------------------------------------------
N <- length(AdoptionSpeed_labels)

# install.packages("xgboost")
library(xgboost)

mod <- xgboost(data = train1, 
               label = train1_label, 
               max.depth = 2, eta = 1, nthread = 2, nrounds = 2, 
               objective = "multi:softprob", 
               num_class= length(AdoptionSpeed_labels))

pred <- matrix(predict(mod, train1),
               nrow= nrow(train1), 
               ncol=length(AdoptionSpeed_labels), 
               byrow=T, 
               dimnames= list(c(),paste0('p',1:N)))

head(apply(pred,1,sum),10)
# pred_label <- apply(pred,1,function(x)which(rmultinom(1,1,x)==1))-1
pred_label <- apply(pred,1,function(x)which(x==max(x)))-1

# training data features and prediction matrix output
train_text <- cbind(train1, pred)
# write.csv(train_text, file='train_text.csv', row.names=F, quote=F)

# accuracy
mean(pred_label==train1_label)

# quadratic weighted kappa
qwk <- function(test_label, pred_label, N){
  # O
  tb <- table(test_label, pred_label)
  O=E=w <- matrix(0, nrow=N, ncol=N, dimnames= list(1:N-1, 1:N-1))
  O[,colnames(tb)] <- tb
  O <- O/sum(O)
  
  # E
  tb <- table(test_label) %o% table(pred_label) 
  E[,colnames(tb)] <- tb
  E <- E/sum(E) 
  
  # kappa
  num=den <- 0
  for(i in 1:N){
    for(j in 1:N){
      w[i,j] <- (i-j)^2/(N-1)^2
      num <- num + w[i,j] * O[i,j]    
      den <- den + w[i,j] * E[i,j]
    }
  }  
  (1-num/den)
}

qwk(train1_label, pred_label, N)


# predict test dataset ----------------------------------------------------

# prepare test dataset
test1 <- test %>% 
  select(text= Description) %>% 
  mutate(line=1:n()) %>% 
  unnest_tokens(word, text) %>% 
  anti_join(stop_words) %>% 
  anti_join(data.frame(word=as.character(1:100))) %>% 
  left_join(lexi, by='word') %>% 
  count(line, sentiment) %>% 
  filter(!is.na(sentiment)) %>% 
  spread(sentiment, n) %>% 
  right_join(data.frame(line=1:nrow(test))) %>% 
  select(-1) %>% 
  as.matrix
test1[is.na(test1)] <- 0

pred <- matrix(predict(mod, test1),
               nrow= nrow(test1), 
               ncol=length(AdoptionSpeed_labels), 
               byrow=T, 
               dimnames= list(c(),paste0('p',1:N)))

# testing data features and prediction matrix output
test_text <- cbind(test1, pred)
# write.csv(test_text, file='test_text.csv', row.names=F, quote=F)

# predict adoption speed category
head(apply(pred,1,sum),10)
pred_label <- apply(pred,1,function(x)which(x==max(x)))-1

sub <- data.frame(PetID= test$PetID, 
                  AdoptionSpeed= pred_label)
write.csv(sub, file='submission.csv', row.names=F, quote=F)

head(sub)
```
